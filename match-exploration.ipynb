### Let's insert column with says, if home team wins, draw or looses

```python
match['winner'] = np.select(condlist=[match.home_team_goal - match.away_team_goal < 0, match.home_team_goal - match.away_team_goal == 0, match.home_team_goal - match.away_team_goal > 0], choicelist=['lost', 'draw', 'win'])
```
